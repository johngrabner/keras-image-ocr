# Transformation from (h,w) to (w, h, 1)

````python
import numpy as np
````

# A numpy array with height first

https://docs.scipy.org/doc/numpy/reference/generated/numpy.array.html

````python
a = np.array([[1,4],[2,5],[3,6]])
print(a)
print(a.shape)
````

# Make width first

https://docs.scipy.org/doc/numpy/reference/generated/numpy.moveaxis.html

````python
b = np.moveaxis(a, -1, 0)
print(b)
print(b.shape)
````

# Add dimension for channel

https://docs.scipy.org/doc/numpy-1.14.1/reference/generated/numpy.expand_dims.html

````python
c = np.expand_dims(b, axis=2)
print(c)
print(c.shape)
````
